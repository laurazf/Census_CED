```python
import requests
import urllib
import base64
import json
from pprint import pprint

class Mendeley:
        
    def __init__(self):
        
        message = "##"        
        message_bytes = message.encode('ascii')
        
        base64_bytes = base64.b64encode(message_bytes)
        base64_message = base64_bytes.decode('ascii')

        header = {'Accept': 'application/json', 
                  'Content-Type': 'application/x-www-form-urlencoded', 
                  'Authorization': 'Basic '+base64_message}

        dat = { 'grant_type': 'client_credentials', 'scope': 'all' }
        url = 'https://api.mendeley.com/oauth/token'
        
        result = requests.post(url, data = dat, headers = header, proxies={"https": "##"})
        
        if (result.status_code == 200):
            self.access = json.loads(result.text)
        else:
            self.access = None

    '''
    Main Search Routine
    '''
    def search(self, _path, _query, _limit=100):
        
        if (self.access is not None):
            header = { 'Accept': 'application/vnd.mendeley-document.1+json', 
                       'Authorization': self.access['token_type']+" "+self.access['access_token'] }

            url = 'https://api.mendeley.com'+_path
        
            result = requests.get(url, params={"query":_query, "limit":_limit}, headers = header, proxies={"https": "http://proxy##"})

            if (result.status_code == 200):
                pprint(json.loads(result.text))
            else:
                print(result.text)

    '''
    Search Catalog for _query string
    '''
    def searchCatalog(self, _query, _limit=100):
        return self.search("/search/catalog", _query, _limit)

'''
Main Program
'''
api = Mendeley()
api.searchCatalog("J Bradford Jensen", 100)

```
